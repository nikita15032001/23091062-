```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(62)
coefficients = np.random.randn(6)

def polynomial(x, coefficients):
    return np.polyval(coefficients, x)

x_values = np.linspace(-5, 5, 50)

y_values = polynomial(x_values, coefficients)

derivative_coefficients = np.polyder(coefficients)
critical_points = np.roots(derivative_coefficients)

valid_critical_points = [x for x in critical_points if -5 <=x<= 5]

y_critical = polynomial(valid_critical_points, coefficients)

plt.plot(x_values, y_values, label='5th Degree Polynomial')
plt.scatter(valid_critical_points, y_critical, color='black', label='Critical Points')

for x,y in zip(valid_critical_points,y_critical):
    label = f'({x:.2f}, {y:.2f})'
    plt.annotate(label, (x, y), textcoords="offset points", xytext=(-10,10), ha='center')

plt.title('Plot of 5th Degree Polynomial along with Maxima and Minima')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
